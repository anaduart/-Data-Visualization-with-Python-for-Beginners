#### Importando biblioteca no Python

*Primeiro Modo*\
Podemos importar toda a biblioteca com todas as suas funções e métodos e apelidando-a.

```python
#importando a biblioteca math e apelidado-a de 'm'
import math as m
```

```python
#usando a função sqrt (raiz quadrada) de 'm'
x = m.sqrt(5)
print(x)
```

*Segundo Modo*\
Podemos também deixar de importar toda a biblioteca para reduzir a memória e importar uma ou mais função específicas da biblioteca (sempre separadas por vígulas), fazendo isso não precisamos apelidar a biblioteca e usar a função importda diretamente no nosso código.

```python
from math import sqrt, cos
```

```python
y = sqrt(5)
print(y)
```

```python
z = cos(90)
print(z)
```

#### Manuseando arquivos no Python

*Primeiro modo*\
Usaremos a função oopen para abrir arquivos com os seguintes argumentos:
* **w**: (write) escreve no arquivo declarado.
* **a**: (add) adiciona caracteres no aquivo declarado.
* **r**: (read) lê o arquivo declarado


```python
# f está habilitado para escreve no arquivo myFile.txt
f = open('myfile.txt', 'w')
```

```python
f.write('Hello')
f.close()
```

```python
# f está habilitado para adicionar caracteres no arquivo myFile.txt
f = open('myfile.txt', 'a')
```

```python
#write escreve apenas strings, talvez é necessário usar a função str()
f.write(' Word')
f.write(' Again')
f.write('\n') #nova linha
f.close()
```

```python
# f está habilitado para ler o arquivo myFile.txt
f = open('myfile.txt', 'r')
line = f.readline()
print(line)
```

```python
#imprimi a linha separadas por linhas vazinha entre uma e outra (\n)
print(line.strip('\n'))
f.close()
```

```python
#separa os elementos do arquivos separados por " "
print(line.strip('\n').split(" "))
```

```python
# é importante converter os números ao escrever nos arquivos
print(int('1'), type(int('1')))
print(str('1'), type(str('1')))
```

```python
# Com esse comando não precisamos fechar o arquivo depois de alguma alteração
with open('myfile.txt', 'r') as f:
    line2=f.readline()
    print(line2)
```

#### Construindo gráficos de linhas e de dispersão

Usaremos a biblioteca matplotlib para contruir nossos gráficos, para isso precisamos importá-lo:

```python
import matplotlib.pyplot as plt
```

```python
#lendo nossos dados
with open('Arquivos\Goals.txt', 'r') as f:
    HomeTeamGoals = [int(x) for x in f.readline().strip('\n').strip(' ').split(' ')]
    AwayTeamGoals = [int(x) for x in f.readline().strip('\n').strip(' ').split(' ')]
```

```python
#gráfico dos gols
plt.plot(HomeTeamGoals)
plt.show()
```

```python
#customizando nossos eixos e acrescentando os outros dados
xVariable=[]
for i in range(len(HomeTeamGoals)):
    xVariable.append(i)
plt.plot(xVariable, AwayTeamGoals, c= 'green')
plt.plot(xVariable, HomeTeamGoals, c= 'blue')
plt.xlim(0,145)
plt.ylim(-0.5,7)
plt.show()
```

```python
#Mesmos dados criando, agora, um gráfico de dispersão
xVariable=[]

for i in range(len(HomeTeamGoals)):
    xVariable.append(i)
plt.scatter(xVariable, AwayTeamGoals, c= 'green', s= 5)
plt.scatter(xVariable, HomeTeamGoals, c= 'blue', s= 5)
plt.xlim(0,145)
plt.ylim(-0.5,7)
plt.show()
```

```python
#Sem sobrepor os dados
xVariable=[]
for i in range(len(HomeTeamGoals)):
    xVariable.append(i)
plt.scatter(xVariable, AwayTeamGoals, c= 'green', s= 5)
plt.show()
plt.scatter(xVariable, HomeTeamGoals, c= 'pink', s= 5)
plt.xlim(0,145)
plt.ylim(-0.5,6)
plt.show()
```

#### Adding Labels, Titles, Axis Ticks, and Changing Line Styles

```python
#podemos usar o $$ para printar fórmulas
xVariable=[]
ticks = ["Game One", "Game Fifty", "Game 100"]
for i in range(len(HomeTeamGoals)):
    xVariable.append(i)
plt.plot(xVariable, AwayTeamGoals, c= 'green', ls = "--")
plt.title(r"Our first plot $\frac{1}{2}$")
plt.plot(xVariable, HomeTeamGoals, c= 'blue')
plt.xlim(0,145)
plt.ylim(-0.5,7)
plt.show()
```

```python
plt.title("Our second plot")
plt.plot(xVariable, HomeTeamGoals, c = "red", ls = ":")
plt.xlim(0,145)
plt.ylim(-0.5,6)
plt.xlabel(r"Game Number$_5$")
plt.ylabel(r"Goals Scored")
plt.xticks([49],["Fourty Nine"])
plt.yticks([3],["Three Goals"])
plt.show()
```

```python
#salvando a figura gerada
plt.figure(figsize = (10,5))
plt.plot(xVariable, AwayTeamGoals, c= 'green', ls = "--", label = "Away Team Goals")
plt.title(r"Our first plot $\frac{1}{2}$")
plt.plot(xVariable, HomeTeamGoals, c= 'blue', ls = ":",  label = "Home Team Goals")
plt.legend( loc = "lower right")
plt.xlim(0,145)
plt.ylim(-1.5,6)
plt.xlabel(r"Game Number$_5$")
plt.ylabel(r"Goals Scored")
plt.xticks([49],["12/12/12"], rotation = 45)
plt.yticks([3],["Three Goals"], rotation = 90)
plt.text (50,4, "novo texto", fontsize = 8, color = 'blue')
plt. annotate ("Text 2", xy = (30,5), xytext = (65,5), arrowprops = dict(facecolor = "green"))
plt.savefig("MyFirstFigure.pdf")
plt.show()
```

#### Creating 1-Dimensional and 2-Dimensional Histograms

```python
#construindo um histograma
#align alinha o eixo x
plt.figure(figsize = (8,5))
plt.hist(HomeTeamGoals, bins = range(8), label = "Home Teams Goals", align = 'left')
plt.legend()
plt.xlabel("Goals Scored")
plt.ylabel("Number of goals scored")
plt.show();
```

```python
#adicionando os outros dados
plt.figure(figsize = (8,5))
plt.hist(x = (HomeTeamGoals, AwayTeamGoals), bins = range(8), 
         label = ["Home Teams Goals", "AwayTeamGoals"], 
         align = 'left')
plt.legend()
plt.xlabel("Goals Scored")
plt.ylabel("Number of goals scored")
plt.show();
```

```python
#adicionando o total de gols
TotalGoals = HomeTeamGoals + AwayTeamGoals 
plt.figure(figsize = (8,5))
plt.hist(x = (HomeTeamGoals, AwayTeamGoals, TotalGoals), bins = range(8), 
         label = ["Home Teams Goals", "AwayTeamGoals", "Total Goals"], 
         align = 'left')
plt.legend()
plt.xlabel("Goals Scored")
plt.ylabel("Number of goals scored")
plt.show();
```

```python
#
plt.hist2d(x = HomeTeamGoals,
           y = AwayTeamGoals, bins = (range(8), range(7)))
plt.xlabel("Home Team Goals")
plt.ylabel("Away Team Goals")
plt.colorbar()
plt.show()
```

#### Changing the Axis Scales

```python
#já estamos em escala linear
x = []
y = []

for i in range (-10000, 10000):
    x.append(i/100)
    y.append(i/100)
    
plt.plot(x,y)
plt.xscale ("linear")
plt.xscale ("linear")
plt.show()
```

```python
#mudando a escala do gráfico para logarítmica
x = []
y = []

for i in range (-10000, 10000):
    x.append(i/100)
    y.append(i/100)
    
plt.plot(x,y)
plt.xscale ("log")
plt.show()
```

```python
#mudando a escala do gráfico para outra base logarítmica
x = []
y = []

for i in range (-10000, 10000):
    x.append(i/100)
    y.append(i/100)
    
plt.plot(x,y)
plt.xscale ("log", basex = 5)
plt.show()
```

```python
#mudando a escala do gráfico, linthreshx = inclina o centro do gráfico
plt.plot(x,y)
plt.xscale ("symlog", basex = 5, subsx = [2,3], linthreshx = 10)
plt.show()
```

```python
#mudando a escala do gráfico para logit
plt.plot(x,y)
plt.xscale ("logit")
plt.show()
```
